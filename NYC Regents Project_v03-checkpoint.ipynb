```python
#research questions
#percent of college ready high school students who took regents
#compare top 5 schools scoring 80 or above on regents to top 5 schools not scoring 80 or above on regents. 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
#drop NA values on CR scores
df=df.replace('na', pd.NA)
df.dropna(subset=['Number Scoring CR', 'Percent Scoring CR'])
df.head(200)
```

```python
#attempting to find how many cells have an 's' value
s_count= (df == 's'). sum()
print (s_count)
```

```python
#replaced values her with lt5
df=df.replace('s', 'lt5')
lt5_count= (df == 'lt5'). sum()
print (lt5_count)
df.head(200)
```

```python
df.shape
```

```python
#adding in new columns to breakdown DBN codes
df['District'] = df['School DBN']. apply (lambda x:x[:2])
df['Borough']= df['School DBN']. apply (lambda x: x[2])
df['School Number']= df['School DBN']. apply (lambda x: x[-3:])
df2= df[['District' , 'Borough' , 'School Number', 'School DBN', 'School Name', 'School Level', 'Year', 'Total Tested', 'Percent Scoring Below 65', 'Percent Scoring 65 or Above' , 'Percent Scoring 80 or Above' , 'Percent Scoring CR']]
df2.head(200)
```

```python
borough_mapping = {
    "M": "Manhattan",
    "X": "Bronx",
    "K": "Brooklyn",
    "Q": "Queens",
    "R": "Staten Island"
}
df2['Borough'] = df2['Borough'].map(borough_mapping)
df2.head(200)
```

```python
#check & and drop lt5 rows
df3= df2.loc[df2['Total Tested'] >= 5]
#df3= df3.drop(df3[df3['Total Tested'] <5], inplace= True, axis= 0)
df3.head(200)
df3.describe()
```

```python
df3.shape
```

```python
df3.dropna(axis='rows', inplace=True)
```

```python
df3.head(200)
```

```python
#Top 5 schools with highest rate of students scoring 80 or above on regents exams
df3.groupby (['School Name']) ['Percent Scoring 80 or Above'].count().sort_values(ascending=False).head().plot(kind='bar')
```

```python
#Top 5 schools with highest rate of students NOT scoring 80 or above on regents exams
df3.groupby (['School Name']) ['Percent Scoring 80 or Above'].count().sort_values(ascending=False).tail().plot(kind='bar')
```

```python
#the schools with the lowest scores are these schools. For each school only one percent of students scored above 80.
df3.groupby (['School Name']) ['Percent Scoring 80 or Above'].count().sort_values().head()
```

```python
#boroughs of top 5 schools scoring 80 or above
df3.groupby (['Borough']) ['Percent Scoring 80 or Above'].count().sort_values(ascending=False).head().plot(kind='bar')
```

```python
#boroughs of top 5 schools NOT scoring 80 or above
df3.groupby (['Borough']) ['Percent Scoring 80 or Above'].count().sort_values(ascending=False).tail().plot(kind='bar')
```
